```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# data frencase teh poci dalam satu bulan
data = {'Jenis Minuman':['Chocolate Milk Tea', 'Apple Tea', 'Mango Tea', 'Orange Tea', 'Original Tea'],'Terjual':[200,50,20,30,250]}
df = pd.DataFrame(data,columns = ['Jenis Minuman','Terjual'])
```

```python
print(df)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plotdata = pd.DataFrame(
    {"Terjual":[200,50,20,30,250]},
    index= ['Chocolate Milk Tea', 'Apple Tea', 'Mango Tea', 'Orange Tea', 'Original Tea'])
plotdata.plot(kind="bar")
plt.title("Penjualan Teh Poci Bulan Agustus")
plt.xlabel("Jenis Minuman")
plt.ylabel("Terjual")
```

```python
plotdata = pd.DataFrame(
    {"Terjual":[200,50,20,30,250]},
    index= ['Chocolate Milk Tea', 'Apple Tea', 'Mango Tea', 'Orange Tea', 'Original Tea'])
plotdata.plot(kind="line",marker="o",color="green",grid="true")
plt.title("Penjualan Teh Poci Bulan Agustus")
plt.xlabel("Jenis Minuman")
plt.ylabel("Terjual")
```

```python
df.groupby(['Jenis Minuman']).sum().plot(kind='pie', subplots=True, shadow="true", startangle=90,figsize=(6,12), autopct='%1.1f%%')
```

```python
Terjual = [200,50,20,30,250]
JenisMinuman = ['Chocolate Milk Tea', 'Apple Tea', 'Mango Tea', 'Orange Tea', 'Original Tea']
plt.scatter(Terjual, JenisMinuman)
plt.title("Penjualan Teh Poci Bulan Agustus")
plt.xlabel("Jenis Minuman")
plt.ylabel("Terjual")
plt.show()
```
